```python
urls = []
for num in range(1, 8):
    url = 'https://www.amazon.com/PlayStation-5-Digital/product-reviews/B09DFHJTF5/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=' + str(num)
    urls.append(url)
```

```python
Reviewers = []
Score = []
Review_text = []

for u in urls:

    driver.get(u)
    time.sleep(3)

    doc = BeautifulSoup(driver.page_source, 'html')
    doc.find(class_ = 'a-section a-spacing-large celwidget').decompose()

    length = len(doc.find_all(class_ = 'a-profile-name'))
    names = [doc.find_all(class_ = 'a-profile-name')[n].text for n in range(0, length)]
    
    for name in names:
        Reviewers.append(name)
    
    ratings = [doc.find_all(class_ = 'a-icon-alt')[n].text for n in range(0, length)]
    num_ratings = [float(ratings[n][0:2]) for n in range(0, length)]

    for r in num_ratings:
        Score.append(r)

    reviews = [doc.find_all(class_ = 'a-row a-spacing-small review-data')[n].text.strip() for n in range(0, length)]

    for review in reviews:
        Review_text.append(review)

driver.quit()

import pandas as pd

df_ps5 = pd.DataFrame({'Reviewers' : Reviewers, 'Score' : Score, 'Review_text' : Review_text})
df_ps5.head()
```
